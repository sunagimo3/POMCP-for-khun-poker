```python
import pandas as pd
import numpy as np

# --- Load file ---
df = pd.read_excel("Cp=0.5.xlsx")

# --- Column index setup ---
error_cols = [2,4,5,6,7,8,9,10,12]
target_values = np.array([0,0,1/3,1,1/3,0,1,0,1])

# Alpha columns
col_x = 1
col_y = 3
col_z = 11

# --- Main squared error ---
vals = df.iloc[:, error_cols].values
err_main = np.sum((vals - target_values)**2, axis=1)

# --- Compute α ---
x = df.iloc[:, col_x].values
y = df.iloc[:, col_y].values
z = df.iloc[:, col_z].values
alpha = (x + 3*y + z - 1/3) / 11

# --- Additional error term ---
add_err = (x - alpha)**2 + (y - 3*alpha)**2 + (z - (1/3) - alpha)**2

# --- Total error ---
df["total_error"] = err_main + add_err

# ---- Extract values ----
total_errors_list = df["total_error"].tolist()

# ==============================
# ① Excel貼り付け用（縦の1列）
# ==============================
print("=== Excel貼り付け（縦1列）===")
for v in total_errors_list:
    print(v)

# ==============================
# ② Excel貼り付け用（横1行）→ セルに展開
# ==============================
print("\n=== Excel貼り付け（横1行）===")
print("\t".join(str(v) for v in total_errors_list))
print(df.columns)
print(df.columns[0])
```
